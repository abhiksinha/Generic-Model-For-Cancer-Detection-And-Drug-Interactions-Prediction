```python
df_inter=pd.read_csv(paths+"drug_drug.tsv",sep='\t')
```

```python
# bortezomib
v=[]
cnt=0
obj={}
#oversampling
sampling_factor=3
for l1,l2 in zip(df_inter['l1'],df_inter['l2']):
    if(l1=="DB00188"):
        for i in range(sampling_factor):
            v.append([l2,1])
        obj[l2]=1
    
for x in df1['drugbank_id']:
    if(x not in obj):
        v.append([x,0])
        cnt+=1;
    if(cnt==500):
        break;
```

```python
df_borte=pd.DataFrame(v,columns=['drugbank_id','interact'])
df_borte.head(10)
```

```python
df_borte_final=pd.merge(df1,df_borte,on='drugbank_id')
df_borte_final
```

```python
import pandas as pd
borte_d=pd.read_csv(paths+"bortezo.tsv", sep='\t')
borte_d=borte_d[['name']]
borte_d.head()
```

```python
df_db=pd.read_csv(paths+"Drugbank.csv")
df_db=df_db[['name','smiles']]
```

```python
df_merge_db= pd.merge(borte_d,df_db,on='name')
df_merge_db.head()
```

```python
df_merge_db.count()
```

```python
df_to_write=df_merge_db[['smiles']]
```

```python
obj={}
for x in df_merge_db['name']:
    obj[x]=1
```

```python
# bortezomib
v=[[x,1] for x in df_to_write['smiles']]
cnt=0
for x,y in zip(df_db['name'],df_db['smiles']):
    if(x not in obj):
        v.append([y,0])
        cnt+=1;
    if(cnt==1000):
        break;
```

```python
df_final_write=pd.DataFrame(v,columns=['smiles','interact'])
df_final_write
```
