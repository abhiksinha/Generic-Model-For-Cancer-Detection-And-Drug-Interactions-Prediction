```python
df_inter.count()
```

```python
# bortezomib
v=[]
cnt=0
obj={}
#oversampling
sampling_factor=3
for l1,l2 in zip(df_inter['l1'],df_inter['l2']):
    if(l1=="DB00188"):
        for i in range(sampling_factor):
            v.append([l2,1])
        obj[l2]=1
    
for x in df1['drugbank_id']:
    if(x not in obj):
        v.append([x,0])
        cnt+=1;
    if(cnt==500):
        break;
```

```python
df_borte=pd.DataFrame(v,columns=['drugbank_id','interact'])
df_borte.head(10)
```

```python
df_borte_final=pd.merge(df1,df_borte,on='drugbank_id')
df_borte_final
```

```python
data_smiles=pd.read_csv("D:\\DRUG\\Drug_CNN\\Drugbank.csv")
data_smiles=data_smiles[['drugbank_id','smiles']]
data_smiles=data_smiles.drop_duplicates()
```

```python
df_borte_smiles=pd.merge(df_borte_final,data_smiles,on='drugbank_id')
```
